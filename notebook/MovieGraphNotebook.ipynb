```python
import pandas as pd
import json, re, json
from pandas.io.json import json_normalize
```

```python
ratings = pd.read_json("movie_ratings_mean.json")
```

```python
movies = pd.read_json("movies_metadata.json")
```

```python
with open('movies_metadata.json') as data_file:    
    movies_json = json.load(data_file)
```

movies_json

```python
with open('credits.json') as data_file:    
    credits_json = json.load(data_file)
```

```python
cast = json_normalize(credits_json, record_path='cast', record_prefix='cast_', meta=['movieId'])
```

```python
cast = cast.astype({'movieId':'int64'})
```

```python
print('Unique movies related to cast', len(pd.unique(cast['movieId'])))
print('Unique cast ids', len(pd.unique(cast['cast_id'])))
print('Unique cast names', len(pd.unique(cast['cast_name'])))
print('Cast count', cast.count())
cast
```

```python
cast.loc[cast.cast_name == 'Tom Hanks']
```

```python
genres = json_normalize(movies_json, record_path='genres', record_prefix='genre_').drop_duplicates()
```

```python
genres.to_json('genres.json', orient='records')
```

```python
collections = json_normalize(movies_json, record_path='belongs_to_collection', record_prefix='collection_')[['collection_id', 'collection_name']].drop_duplicates()
```

```python
collections.to_json('collections.json', orient='records')
```

```python
movies_for_cast = movies[['id', 'revenue', 'genres']]
movies_for_cast = movies_for_cast.assign(genres=movies_for_cast.genres.map(lambda x: [item['id'] for item in x]))
```

```python
ratings = pd.read_csv('ratings.csv').groupby(['movieId'])['rating'].mean()
```

```python
movies_for_cast.dtypes
```

```python
movies_with_ratings_for_cast = movies_for_cast.merge(ratings, how='left', left_on='id', right_on='movieId').fillna(0)
```

```python
cast_with_movie_data = cast.merge(movies_with_ratings_for_cast, how='left', left_on='movieId', right_on='id')
cast_with_movie_data.dtypes
```

```python
cast_with_movie_data
```

```python
cast_to_export = cast_with_movie_data.groupby(['cast_id', 'cast_name']).agg({'revenue': 'sum',
                                                            'rating': 'mean',
                                                            'movieId': lambda x: list(x),
                                                            'cast_order': lambda x: list(set(x)),
                                                            'genres': lambda x: list(set([item for sublist in x for item in sublist]))}).reset_index()
cast_to_export
```

```python
cast_to_export = cast_to_export.rename(columns={'cast_id': 'talent_id', 'cast_name': 'name', 'movieId': 'movies', 'cast_order': 'role_order'})
```

```python
cast_to_export['job'] = 'Actor'
```

```python
cast_to_export['job'] = cast_to_export['job'].map(lambda x: [x])
```

```python
cast_to_export
```

```python
cast_to_export.to_json("cast_with_movie.json", orient='records')
```

```python
crew = json_normalize(credits_json, record_path='crew', record_prefix='crew_', meta=['movieId'])
crew = crew.astype({'movieId':'int64'})
```

```python
crew.loc[crew.crew_name == 'Tom Hanks']
```

```python
crew_with_movie_data = crew.merge(movies_with_ratings_for_cast, how='left', left_on='movieId', right_on='id')
crew_with_movie_data.dtypes
```

```python
crew_to_export = crew_with_movie_data.groupby(['crew_id', 'crew_name']).agg({'revenue': 'sum',
                                                    'rating': 'mean',
                                                    'movieId': lambda x: list(set(x)),
                                                    'genres': lambda x: list(set([item for sublist in x for item in sublist])),
                                                    'crew_job': lambda x: list(set(x))
                                                }).reset_index()

crew_to_export
```

```python
crew_to_export = crew_to_export.rename(columns={'crew_id': 'talent_id', 'crew_name': 'name', 'movieId': 'movies', 'crew_job': 'job'})
```

```python
crew_to_export.insert(5, 'role_order', 0)
```

```python
crew_to_export['role_order'] = crew_to_export['role_order'].map(lambda x: [x])
```

```python
crew_to_export
```

```python
# Union of crew and cast
talent_union = pd.concat([cast_to_export,crew_to_export])
```

```python
talent = talent_union.groupby(['talent_id', 'name']).agg({'revenue': 'sum',
                                                    'rating': 'mean',
                                                    'movies': lambda x: list(set([item for sublist in x for item in sublist])),
                                                    'role_order': lambda x: list(set([item for sublist in x for item in sublist])),
                                                    'genres': lambda x: list(set([item for sublist in x for item in sublist])),
                                                    'job': lambda x: list(set([item for sublist in x for item in sublist]))
                                                }).reset_index()
talent
```

```python
# Export talent to json
talent.to_json('talent.json', orient='records')
```

```python
talent[:1000].to_json('talent1000.json', orient='records')
```

```python
movies_reduced = movies[['id', 'title', 'budget', 'belongs_to_collection']]
```

```python
movies_reduced
```

```python
movies_reduced = movies_reduced.assign(collections=movies_reduced['belongs_to_collection'].map(lambda x: [i['id'] for i in x]))
```

```python
movies_reduced = movies_reduced.merge(ratings, how='left', left_on='id', right_on='movieId').fillna(0)
```

```python
movies_reduced['efficiency'] = movies_reduced['rating']/movies_reduced['budget']
```

```python
movies_to_export = movies_reduced[['id', 'title', 'budget', 'rating', 'collections']]
```

```python
movies_to_export.to_json('movies.json', orient='records')
```
